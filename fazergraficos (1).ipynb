```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4]
y = [2,3,4,3]

plt.plot(x,y)
plt.show()

# axis = eixo x e y
# axes = quadrante do grafico
# label = legenda
```

```python
plt.plot(x,y,label='dados')
plt.ylabel('Eixo y')
plt.xlabel('Eixo x')
plt.title('Titulo do grafico')
#plt.xticks([0,2,4,6,8,9])
#plt.yticks([-1,3,5,9,11])
plt.legend()
plt.show()
```

```python
plt.plot(x,y, linestyle='dashed', color='r', lw=10.0)
plt.show()
#dá para buscar cores na net
#lw = largura do pontilhado
```

```python
# figura.add_subplot(131) = posição do grafic

valoresx = [1, 2, 3,4]
valoresy = [1, 4, 2, 3]

figura = plt.figure(figsize=(20,3))
figura.suptitle('Titulo geral')

figura.add_subplot(131)
plt.plot(valoresx, valoresy, label='Um dado')
plt.legend()
plt.title('Grafico 1')

figura.add_subplot(133)
plt.scatter(valoresx, valoresy)
plt.title('Grafico 3')

figura.add_subplot(132)
plt.bar(valoresx, valoresy)
plt.title('Grafico 2')

plt.show()
```

```python
#EXERCICIO 1 - Usando os dados do site https://www.kaggle.com/datasets/mattiuzc/stock-exchange-data?resource=download, 

import pandas as pd

data_cotacao = pd.read_csv('indexData.csv.zip')
display(data_cotacao)

data_cotacao1 = data_cotacao[['Index', 'Date', 'Close']]
display(data_cotacao1)

dfremove = data_cotacao1.loc[(data_cotacao1['Index']!='NYA')]
data_cotacaofinal = data_cotacao1.drop(dfremove.index)
display(data_cotacaofinal)

cotacao_final = data_cotacaofinal[13900:]
display(cotacao_final)

plt.plot(cotacao_final['Date'], cotacao_final['Close'], label='NYA', color='g', lw='2')
plt.legend( loc=2, fontsize=10)
plt.ylabel('Valor do fechamento')
plt.xlabel('data da cotação')
plt.title('grafico de cotação da bolsa de NYA')
plt.show()
```
